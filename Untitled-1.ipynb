```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-1,1,100)
y=x**2

plt.figure(figsize=(5,3))
plt.plot(x,y, label='y=x^2')
plt.legend()
plt.show()
```

```python
import numpy as np

score=np.array([71,88,60,90,65])
result=np.where(score>=70,'pass','fail')
print(result)
```

```python
import numpy as np

score=np.array([71,80,68,90,65])
result=np.where(score>=90,'A',
                np.where(score>=80,'B',
                         np.where(score>=70,'C','D')))
print(result)
```

```python
score=[71,80,60,90,65]
result=['pass' if a>=70 else 'fail' for a in score]
print(result)
```
